```python
with open("input.txt", "r") as f:
    data = f.read()

print(data)
```

```python
floor = 0
for char in data:
    if char == "(":
        floor += 1
    elif char == ")":
        floor -= 1
print(floor)
```

```python
floor = 0
for i, char in enumerate(data):
    if char == "(":
        floor += 1
    elif char == ")":
        floor -= 1
    if floor == -1:
        print(i + 1)
        break
```
