```python
with open("input.txt", "r") as f:
    dirs = f.read()

print(dirs)
```

```python
visited = set([(0, 0)])
x, y = 0, 0
for dir in dirs:
    if dir == "^":
        y += 1
    elif dir == "v":
        y -= 1
    elif dir == ">":
        x += 1
    elif dir == "<":
        x -= 1
    visited.add((x, y))
print(len(visited))
```

# Part 2

```python
visited = set([(0, 0)])
s_x, s_y, r_x, r_y = 0, 0, 0, 0
for i, dir in enumerate(dirs):
    if i % 2 == 0:
        x, y = s_x, s_y
    else:
        x, y = r_x, r_y
    
    if dir == "^":
        y += 1
    elif dir == "v":
        y -= 1
    elif dir == ">":
        x += 1
    elif dir == "<":
        x -= 1

    visited.add((x, y))

    if i % 2 == 0:
        s_x, s_y = x, y
    else:
        r_x, r_y = x, y
    
print(len(visited))
```
